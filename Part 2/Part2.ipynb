## Final Project: Smart Quiz Bot

```python
# 1. DATA (List of Dictionaries)
quiz_data = [
    {"question": "Python ka creator kaun hai?", "answer": "Guido"},
    {"question": "AI ka full form kya hai?", "answer": "Artificial Intelligence"},
    {"question": "Google ka AI model kaunsa hai?", "answer": "Gemini"}
]

score = 0

print("--- WELCOME TO AI QUIZ ---")

for item in quiz_data:
    print("Sawal: ",{item['question']})
    user_ans = input("Jawab: ")
    
    # 3. LOGIC (Checking)
    if user_ans.lower() == item['answer'].lower():
        print("Sahi Jawab!")
        score = score + 1
    else:
        print("Ghalat Jawab!")

# 4. RESULT
print(f"Game Over! Aapka Final Score: {score}/{len(quiz_data)}")
```
